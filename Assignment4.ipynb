```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.style.use('seaborn-colorblind')
```

```python
df = pd.read_excel('PopulationWorldBank.xlsx',index_col='Country Name')
```

```python
df.head()
```

```python
#Enter the country name
country = input("Enter a country name ")
Start = int(input("Enter Start Year "))
End = int(input("Enter End Year "))

Syear = Start - 1960 #1960 is the starting year in a given data set
Eyear = End - 1960 #1960 is the starting year in a given data set

#selecting number of years for which graph is to be drawn
years = df.iloc[0:,Syear+3:Eyear+4]

#Taking all the years in a dataframe
year = []
for i in df.iloc[0:,Syear+3:Eyear+4]:
    year.append(i)
#Find the average of population growth(annual %) of all countries
average = []
for i in years:
    avg = df[i].sum()/263
    average.append(avg)

#Taking all data of the country which we entered    
ind_data = []
for i in years:
    ind_data.append(df.loc[country][i]) 

#Plotting a graph of given country with respect to world.
plt.figure()
plt.plot(year,average,'-o',label='World Rate')
plt.plot(year,ind_data,'-o',label='{}'.format(country)+' Rate')
plt.xticks(rotation=45)
plt.title('Population Growth rate of World Vs Population Growth rate of {}'.format(country))
plt.xlabel('Years')
plt.ylabel('Population Growth (annual %)')
plt.legend()
```
